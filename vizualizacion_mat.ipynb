```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
#coniguración para mostrar graficos
#solo funciona en jupyter
%matplotlib inline 

#configuracion para mejorar la calidad de los graficos
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100
```

GRÁFICOS DE LÍNEAS

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y = np.sin(x) # Seno de los puntos

#crear grafico
plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.title('Función de Seno')
plt.xlabel('x')
plt.ylabel('seno(x)')
plt.grid(True) # Añadir una cuadrícula
plt.legend(['Seno']) # Añadir leyenda
plt.show() # Mostrar el grafico
```

GRÁFICO DE DISPERSIÓN 

- Ideal para mostrar relacion entre dos variables y detectar correlaciones

```python
# diagrama de dispersión
np.random.seed(42) # Para reproducibilidad
x = np.random.rand(50) # 100 puntos aleatorios en x
y = x + np.random.normal(0,0.2,50) # y = x + ruido aleatorio
plt.figure(figsize=(8,4))
plt.scatter(x, y, c='blue', alpha=0.5) # Graficar puntos
plt.title('Diagrama de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True, linestyle='--', alpha= .7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

GRÁFICOS DE BARRAS

- Excelentes para comparar categorías o grupos discretos

```python
# Datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 33]
# Crear un gráfico de barras
plt.figure(figsize=(8,4))
plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

HITOGRAMA

- Muestran la distribución  de una varriable continua, dividiendo los datos en intervalos (bins)

```python
#datos
np.random.seed(42) # Para reproducibilidad
datos = np.random.normal(0,1, 1000) # 1000 puntos aleatorios de una normal

# gráfico
plt.figure(figsize=(8,4))
plt.hist(datos, bins=30, color='orange', alpha=0.7) # Histograma
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico

```

GRÁFICO CIRCULAR( PIE CHART)

- Son útiles para mostrar proporciones o porcentajes de un todo

```python
# datos
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]
# Crear un gráfico de pastel
plt.figure(figsize=(8,4))
plt.pie(valores, labels=categorias, shadow=True,
        autopct='%1.1f%%', startangle=90, 
        colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gráfico de Pastel')
plt.axis('equal') # Para que el gráfico sea un círculo
plt.show() # Mostrar el grafico
```

BOXPLOT (DIAGRAMA DE CAJA)

- Muestran la distribución  d elos datos a travez de sus cuartiles, permitiendo identificar outliers

```python
# datos
np.random.seed(42) # Para reproducibilidad
datos = [np.random.normal(0, std, 100) for std in range(1,5)] # 100 puntos aleatorios de una normal

# Crear un gráfico de caja
plt.figure(figsize=(8,6))
plt.boxplot(datos, vert=True, patch_artist=True)
plt.title('Gráfico de Caja')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

COLORES, ESTILOS DE LÍNEA Y MARCADORES

SUBPLOT. MULTIPLES GRAFICOS EN UNA FIGURA

```python
# datos
x= np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y1 = np.sin(x) # Seno de los puntos
y2 = np.cos(x) # Coseno de los puntos
y3 = np.sin(x)*np.cos(x) # Seno por Coseno
y4 = np.sin(x)*  np.exp(-x/5) # Seno por exponencial decreciente

# figura 2x2 subplots
plt.figure(figsize=(10, 8))

# subplot 1: seno

plt.subplot(2, 2, 1) # fila 2, columna 2, indifce 1
plt.plot(x, y1, 'b-')
plt.title('Función Seno')
plt.grid(True) # Añadir una cuadrícula

# subplot 2: coseno
plt.subplot(2, 2, 2) # fila 2, columna 2, indifce 2
plt.plot(x, y2, 'r-')
plt.title('Función Coseno')
plt.grid(True) # Añadir una cuadrícula

# subplot 3: seno por coseno
plt.subplot(2, 2, 3) # fila 2, columna 2, indifce 3
plt.plot(x, y3, 'g-')
plt.title('Seno por Coseno')
plt.grid(True) # Añadir una cuadrícula

# subplot 4: seno por exponencial decreciente
plt.subplot(2, 2, 4) # fila 2, columna 2, indifce 4
plt.plot(x, y4, 'm-')
plt.title('Seno por Exponencial Decreciente')
plt.grid(True) # Añadir una cuadrícula

plt.tight_layout() # Ajustar el espaciado
plt.show() # Mostrar el grafico
```

CASO PRACTICO

```python
# DATOS SIMULADOS
np.random.seed(42) # Para reproducibilidad

# meses

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
    
# productos
categorias = ['Electrónica', 'Ropa', 'Hogar', 'Deportes']

# regiones
regiones = ['Norte', 'Sur', 'Este', 'Oeste']

# datframe
datos = []

for mes in meses:
    for categoria in categorias:
        for region in regiones:
            # ventas mensuales con tendencia estacional
            mes_indice = meses.index(mes)
            base_ventas = 1000 + 500 * np.sin(mes_indice / 12 * 2 * np.pi)
            
            #variacion por categoria
            categoria_factor =  1.3 if categoria == 'Electrónica' else \
                                1.2 if categoria == 'Ropa' else \
                                0.9 if categoria == 'Hogar' else 0.7 
            
            # variacion por region
            region_factor = 1.3 if region == 'Norte' else \
                            1.1 if region == 'Centro' else \
                            0.9 if region == 'Sur' else \
                            1.0 if region == 'Este' else 0.8
                            
            # ventas con ruido aleatorio
            ventas = base_ventas * categoria_factor * region_factor * ( 1 + np.random.normal(0, 0.1))
            
            # ventas diarias promedio
            dias_mes = 30
            ventas_diarias = ventas / dias_mes
            
            # agregar al dataframe
            datos.append({
                'Mes': mes,
                'Categoria': categoria,
                'Region': region,
                'Ventas': ventas,
                'Ventas_Diarias': ventas_diarias
            })
            
# crear dataframe
df = pd.DataFrame(datos)
 
 # mostrar las primeras filas
df.head(15)
             
                
```

```python
# guardar el dataframe
df.to_csv('data_ventas.csv', index=False) # Guardar como CSV
print('datos guardados en data_ventas como CSV.\n')
```
