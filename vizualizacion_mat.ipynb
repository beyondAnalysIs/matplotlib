```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
#coniguración para mostrar graficos
#solo funciona en jupyter
%matplotlib inline 

#configuracion para mejorar la calidad de los graficos
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100
```

GRÁFICOS DE LÍNEAS

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y = np.sin(x) # Seno de los puntos

#crear grafico
plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.title('Función de Seno')
plt.xlabel('x')
plt.ylabel('seno(x)')
plt.grid(True) # Añadir una cuadrícula
plt.legend(['Seno']) # Añadir leyenda
plt.show() # Mostrar el grafico
```

GRÁFICO DE DISPERSIÓN 

- Ideal para mostrar relacion entre dos variables y detectar correlaciones

```python
# diagrama de dispersión
np.random.seed(42) # Para reproducibilidad
x = np.random.rand(50) # 100 puntos aleatorios en x
y = x + np.random.normal(0,0.2,50) # y = x + ruido aleatorio
plt.figure(figsize=(8,4))
plt.scatter(x, y, c='blue', alpha=0.5) # Graficar puntos
plt.title('Diagrama de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True, linestyle='--', alpha= .7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

GRÁFICOS DE BARRAS

- Excelentes para comparar categorías o grupos discretos

```python
# Datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 33]
# Crear un gráfico de barras
plt.figure(figsize=(8,4))
plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

HITOGRAMA

- Muestran la distribución  de una varriable continua, dividiendo los datos en intervalos (bins)

```python
#datos
np.random.seed(42) # Para reproducibilidad
datos = np.random.normal(0,1, 1000) # 1000 puntos aleatorios de una normal

# gráfico
plt.figure(figsize=(8,4))
plt.hist(datos, bins=30, color='orange', alpha=0.7) # Histograma
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico

```

GRÁFICO CIRCULAR( PIE CHART)

- Son útiles para mostrar proporciones o porcentajes de un todo

```python
# datos
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]
# Crear un gráfico de pastel
plt.figure(figsize=(8,4))
plt.pie(valores, labels=categorias, shadow=True,
        autopct='%1.1f%%', startangle=90, 
        colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gráfico de Pastel')
plt.axis('equal') # Para que el gráfico sea un círculo
plt.show() # Mostrar el grafico
```

BOXPLOT (DIAGRAMA DE CAJA)

- Muestran la distribución  d elos datos a travez de sus cuartiles, permitiendo identificar outliers

```python
# datos
np.random.seed(42) # Para reproducibilidad
datos = [np.random.normal(0, std, 100) for std in range(1,5)] # 100 puntos aleatorios de una normal

# Crear un gráfico de caja
plt.figure(figsize=(8,6))
plt.boxplot(datos, vert=True, patch_artist=True)
plt.title('Gráfico de Caja')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

COLORES, ESTILOS DE LÍNEA Y MARCADORES

```python
# datos 
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y1 = np.sin(x) # Seno de los puntos
y2 = np.cos(x) # Coseno de los puntos

# Crear un gráfico con múltiples líneas
plt.figure(figsize=(10,6))

# diferentes colores y estilos de línea
plt.plot(x, y1, color='blue', linestyle='-', marker='o', 
         linewidth=2, label='Seno(x)')
plt.plot(x, y2, color='red', linestyle='--', marker='^',
         linewidth=2, label='Coseno(x)')

plt.title('Gráfico de Líneas')
plt.xlabel('x')
plt.ylabel('Valores')
plt.legend() # Añadir leyenda
plt.grid(True) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

PERSONALIZACIÓN DE EJES

```python
# datos
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y = x**2 # Cuadrado de los puntos

# grafico
plt.figure(figsize=(8,6))
plt.plot(x, y, 'g-') # 'g-' es el color verde
plt.title('Personalización de Ejes')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#personalizar los ejes
plt.xlim(0, 8) # Limitar el eje x
plt.ylim(0, 50) # Limitar el eje y
plt.xticks([0, 2, 4, 6, 8], ['0', '2', '4', '6', '8']) # Cambiar los ticks del eje x
plt.yticks([0,25,50], ['Bajo', 'Medio', 'Alto']) # Cambiar los ticks del eje y

plt.grid(True) # Añadir una cuadrícula
plt.show() # Mostrar el grafico
```

AÑADIR TEXTO Y ANOTACIONES

```python
# DATOS
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y = np.sin(x) # Seno de los puntos

# gráfico
plt.figure(figsize=(10,6))
plt.plot(x , y, 'b-')
plt.title('Apuntes de Texto')
plt.xlabel('X')
plt.ylabel('Seno(X)')

# añadir texto
plt.text(4, 0.8, 'Máximo', fontsize=12, color='red')

# añadir anotación con flecha
plt.annotate('Minimo local',
             xy=(4.65, -0.99), # coordenadas del punto
             xytext=(3.8, -0.5), # posición del texto
             arrowprops=dict(facecolor='black', shrink=0.05)) # propiedades de la flecha
plt.grid(True) # Añadir una cuadrícula
plt.show() # Mostrar el grafico

```

ESTILOS 

```python
# estilos  disponibles
print('estilos disponibles:\n')
print(plt.style.available)
```

```python
# datos
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y = np.sin(x) # Seno de los puntos

# gráfico
estilos = ['ggplot', 'seaborn-v0_8-darkgrid', 'bmh', 'dark_background']

# verificar que estilos existen
estilos_disponibles = [estilo for estilo in estilos if estilo in plt.style.available]
if not estilos_disponibles:
    estilos_disponibles = plt.style.available[:5]
    
#mostrar el mismo grafico con diferentes estilos
plt.figure(figsize=(15,10))

for i, estilo in enumerate(estilos_disponibles,1):
    with plt.style.context(estilo):
        ax = plt.subplot(2, 2, i)
        ax.plot(x, y)
        ax.set_title(f'Estilo: {estilo}')
        ax.set_xlabel('x')

plt.tight_layout() # Ajustar el espaciado
plt.show() # Mostrar el grafico
```
